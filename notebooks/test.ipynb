```python
from numpyword2vec.word2vec import Word2Vec
from datasets import load_dataset
```

```python
def load_text8_from_hf(split="train", sentence_len=1000):

    ds = load_dataset("afmck/text8", split=split)

    text = ds[0]["text"]
    tokens = text.split()

    corpus = []
    for i in range(0, len(tokens), sentence_len):
        chunk = tokens[i : i + sentence_len]
        if len(chunk) > 1:
            corpus.append(chunk)

    return corpus
```
